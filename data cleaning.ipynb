# genres

```python
df.genres.values[0]
```

```python
ast.literal_eval(df.genres.values[0])
```

```python
UniqueGenres = set([])

for i in df.genres:
    genlist = ast.literal_eval(i)
    for j in genlist:
        UniqueGenres.add(j['name'])
```

```python
UniqueGenres
```

```python
rows = [[]]
genres = list(UniqueGenres)

for i in df.genres:
    genlist = ast.literal_eval(i)
    moviesgen = [x['name'] for  x in genlist]
    row = []
    for i in genres:
        row.append(int(i in moviesgen))
    rows.append(row)
```

```python
rows
```

```python
genre = pd.DataFrame(rows[1:],columns= genres)
genre
```

```python
df = pd.concat([df,genre], axis = 1)
```

```python
df
```

```python
df = df.drop(columns= ['genres'])
```

```python
df
```
